```python
# Import NLTK library
import nltk

# Download tokenizer data (only first time)
nltk.download('punkt')

# Import word tokenizer
from nltk.tokenize import word_tokenize

# Input text
text = "I am learning Natural Language Processing."

# Convert sentence into words
tokens = word_tokenize(text)

# Print tokens
print(tokens)
```

```python
import nltk
nltk.download('punkt')

# Import sentence tokenizer
from nltk.tokenize import sent_tokenize

# Multiple sentences
text = """
I love AI.
NLP is very interesting.
I want to become an AI Engineer.
"""

# Split paragraph into sentences
sentences = sent_tokenize(text)

# Print sentences
print(sentences)
```

```python
# Input text
text = "I am learning NLP"

# Split on spaces
tokens = text.split()

# Print result
print(tokens)
```

```python
# Import regex module
import re

# Input text
text = "Hello! I am learning NLP, version 2.0."

# \w+ means extract only words and numbers
tokens = re.findall(r'\w+', text)

# Print tokens
print(tokens)
```

```python
# Import tokenizer
from nltk.tokenize import TreebankWordTokenizer

# Create object
tokenizer = TreebankWordTokenizer()

# Input text
text = "I'm learning NLP."

# Tokenize text
tokens = tokenizer.tokenize(text)

# Print result
print(tokens)
```

```python
# Import tokenizer
from nltk.tokenize import TweetTokenizer

# Create tokenizer object
tokenizer = TweetTokenizer()

# Input tweet
text = "I love #AI 😊 @OpenAI"

# Tokenize tweet
tokens = tokenizer.tokenize(text)

# Print tokens
print(tokens)
```

```python
# Import tokenizer
from nltk.tokenize import WordPunctTokenizer

# Create object
tokenizer = WordPunctTokenizer()

# Input text
text = "I am learning NLP, version 2.0."

# Tokenize
tokens = tokenizer.tokenize(text)

# Print tokens
print(tokens)
```

```python
# Import tokenizer
from nltk.tokenize import WhitespaceTokenizer

# Create object
tokenizer = WhitespaceTokenizer()

# Input text
text = "I am learning NLP"

# Split based on spaces only
tokens = tokenizer.tokenize(text)

# Print tokens
print(tokens)
```

```python
# Import spaCy
import spacy

# Load English language model
nlp = spacy.load("en_core_web_sm")

# Input text
text = "I am learning Natural Language Processing."

# Convert text into spaCy document
doc = nlp(text)

# Print each token
for token in doc:
    print(token.text)
```

```python
# Import BERT tokenizer
from transformers import BertTokenizer

# Download pretrained tokenizer
tokenizer = BertTokenizer.from_pretrained(
    "bert-base-uncased"
)

# Input text
text = "I am learning NLP."

# Tokenize sentence
tokens = tokenizer.tokenize(text)

# Print tokens
print(tokens)
```

```python
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained(
    "bert-base-uncased"
)

text = "I love NLP"

# Convert words into token IDs
ids = tokenizer.encode(text)

# Print IDs
print(ids)
```

```python
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained(
    "bert-base-uncased"
)

text = "I love NLP"

# Encode sentence
encoding = tokenizer(text)

# Print dictionary
print(encoding)
```

```python
from nltk.tokenize import RegexpTokenizer

# Extract only words (ignore punctuation)
tokenizer = RegexpTokenizer(r'\w+')

text = "Hello! I am learning NLP."

tokens = tokenizer.tokenize(text)

print(tokens)
```
